Eliminar valores nulos

```python
def show_null(df):
    total = df.isnull().sum().sort_values(ascending=False)
    percent = (df.isnull().sum()/df.isnull().count()).sort_values(ascending=False)
    missing_data = pd.concat([total, percent*100], axis=1, keys=['Valores nulos totales', 'Valores nulos %'])
    display(missing_data)
show_null(df)
```

```python
df = df.dropna()
show_null(df)
```

```python
df
```

Definir X e Y
 - Y: RainTomorrow
 - X: Todos los demás valores

```python
X_df = df.drop(columns=['RainTomorrow'])
X_df
```

```python
X = X_df.to_numpy()
X
```

```python
Y_df = df.get("RainTomorrow")
Y_df
```

```python
Y = Y_df.to_numpy()
Y = Y.reshape(-1,1)
Y
```

```python
from sklearn.preprocessing import LabelEncoder
le1 = LabelEncoder()
X[:,0] = le1.fit_transform(X[:,0])
le2 = LabelEncoder()
X[:,6] = le2.fit_transform(X[:,6])
le3 = LabelEncoder()
X[:,8] = le3.fit_transform(X[:,8])
le4 = LabelEncoder()
X[:,9] = le4.fit_transform(X[:,9])
le5 = LabelEncoder()
X[:,-1] = le5.fit_transform(X[:,-1])
le6 = LabelEncoder()
Y[:,-1] = le6.fit_transform(Y[:,-1])
```

```python
X
```

```python
Y = np.array(Y,dtype=float)
Y
```

Aplicar RandomForestClassifier

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators=100,random_state=0)
classifier.fit(X_train,Y_train)
```

```python
classifier.score(X_train,Y_train)
```

```python
y_pred = le6.inverse_transform(np.array(classifier.predict(X_test),dtype=int))
Y_test = le6.inverse_transform(np.array(Y_test,dtype=int))
```

```python
y_pred = y_pred.reshape(-1,1)
Y_test = Y_test.reshape(-1,1)
y_pred = y_pred.astype(int)
Y_test = Y_test.astype(int)
```

```python
df2 = np.concatenate((Y_test,y_pred),axis=1)
dataframe = pd.DataFrame(df2,columns=['Rain on Tommorrow','Predition of Rain'])
dataframe
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(Y_test,y_pred)
```
